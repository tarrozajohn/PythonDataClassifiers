# 1) Load Data

```python
import csv
import numpy as np
import matplotlib as plt
import pandas as pd
from sklearn import datasets, linear_model, naive_bayes, neighbors, svm

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

# load in data
header_list = ['x1', 'x2', 'y']
data1 = pd.read_csv("dataset1.csv", delimiter= ",", quoting=csv.QUOTE_NONE, header = None, names = header_list)
data2 = pd.read_csv("dataset2.csv", delimiter= ",", quoting=csv.QUOTE_NONE, header = None, names = header_list)
data3 = pd.read_csv("dataset3.csv", delimiter= ",", quoting=csv.QUOTE_NONE, header = None, names = header_list)

# observes first five data points of each dataset
data1View = data1.head()
data2View = data2.head()
data3View = data3.head()

print("Data 1:\n", data1View, '\n')
print("Data 2:\n", data2View, '\n') 
print("Data 3:\n", data3View, '\n')
```

# 2) Fit and Score data1

```python
# Set features and target for Dataset1
feature_cols = ['x1', 'x2']
X = data1[feature_cols] # Features
y = data1.y # Target variable

# linear regresssion:
lr_model = linear_model.LogisticRegression()
# Fit data1   
lr_model.fit(X, y)
# Score data1  
lr_data1_train_score = lr_model.score(X, y)

# Gaussian naive Bayes:
gnb_model = naive_bayes.GaussianNB()
# Fit data1 using 
gnb_model.fit(X, y)
# Score data1 using Gaussian naive Bayes
gnb_model_train_score = gnb_model.score(X, y)

# KNeighbors Classifier:
knn_model = neighbors.KNeighborsClassifier()
# Fit data1 using 
knn_model.fit(X, y)
# Score data1 using Gaussian naive Bayes
knn_model_train_score = knn_model.score(X, y)

# Print results:
print("Linear Regresssion Score: ", lr_data1_train_score, "\n")
print("Gaussian Naive Bayes Score: ", gnb_model_train_score, "\n")
print("KNeighbors Classifier Score: ", knn_model_train_score, "\n")
```

How accurate are the model’s predictions?

The model’s predictions were accurate since training with the whole data and testing on the whole data

# 6) Plot the decision boundaries for each classifier and dataset

```python
# define bounds of the domain
# data1
data1_minX1, data1_maxX1 = data1.iloc[:, 0].min()-1, data1.iloc[:, 0].max()+1
data1_minX2, data1_maxX2 = data1.iloc[:, 1].min()-1, data1.iloc[:, 1].max()+1
# data2
data2_minX1, data2_maxX1 = data2.iloc[:, 0].min()-1, data2.iloc[:, 0].max()+1
data2_minX2, data2_maxX2 = data2.iloc[:, 1].min()-1, data2.iloc[:, 1].max()+1
# data3
data3_minX1, data3_maxX1 = data3.iloc[:, 0].min()-1, data3.iloc[:, 0].max()+1
data3_minX2, data3_maxX2 = data3.iloc[:, 1].min()-1, data3.iloc[:, 1].max()+1

# define the x and y scale
# data1
data1_x1grid = np.arange(data1_minX1, data1_maxX1, 0.1)
data1_x2grid = np.arange(data1_minX2, data1_maxX2, 0.1)
# data2
data2_x1grid = np.arange(data2_minX1, data2_maxX1, 0.1)
data2_x2grid = np.arange(data2_minX2, data2_maxX2, 0.1)
# data3
data3_x1grid = np.arange(data3_minX1, data3_maxX1, 0.1)
data3_x2grid = np.arange(data3_minX2, data3_maxX2, 0.1)

# create all of the lines and rows of the grid
# data1
data1_xx, data1_yy = np.meshgrid(data1_x1grid, data1_x2grid)
# data2
data2_xx, data2_yy = np.meshgrid(data2_x1grid, data2_x2grid)
# data3
data3_xx, data3_yy = np.meshgrid(data3_x1grid, data3_x2grid)

# flatten each grid to a vector
# data1
data1_r1, data1_r2 = data1_xx.flatten(), data1_yy.flatten()
data1_r1, data1_r2 = data1_r1.reshape((len(data1_r1), 1)), data1_r2.reshape((len(data1_r2), 1))
# data2
data2_r1, data2_r2 = data2_xx.flatten(), data2_yy.flatten()
data2_r1, data2_r2 = data2_r1.reshape((len(data2_r1), 1)), data2_r2.reshape((len(data2_r2), 1))
# data3
data3_r1, data3_r2 = data3_xx.flatten(), data3_yy.flatten()
data3_r1, data3_r2 = data3_r1.reshape((len(data3_r1), 1)), data3_r2.reshape((len(data3_r2), 1))

# horizontal stack vectors to create x1,x2 input for the model
# data1
data1_grid = np.hstack((data1_r1,data1_r2))
# data2
data2_grid = np.hstack((data2_r1,data2_r2))
# data3
data3_grid = np.hstack((data3_r1,data3_r2))

# Set features 
feature_cols = ['x1', 'x2']

# LINEAR REGRESSION:
lr_model = linear_model.LogisticRegression()

# make predictions for the grid (WARNING HERE)
# Fit data1   
lr_model.fit(data1[feature_cols].values, data1.y)
# data1 predictions
data1_yhat = lr_model.predict(data1_grid)
# Fit data2   
lr_model.fit(data2[feature_cols].values, data2.y)
# data2 predictions
data2_yhat = lr_model.predict(data2_grid)
# Fit data3   
lr_model.fit(data3[feature_cols].values, data3.y)
# data3 predictions
data3_yhat = lr_model.predict(data3_grid)

# reshape the predictions back into a grid
# data1
data1_zz = data1_yhat.reshape(data1_xx.shape)
# data2
data2_zz = data2_yhat.reshape(data2_xx.shape)
# data3
data3_zz = data3_yhat.reshape(data3_xx.shape)

# Print contour graphs
fig, (ax1, ax2, ax3) = plt.pyplot.subplots(1, 3)
print("Linear Regression:")

# data1
data1_graph = plt.pyplot.figure(1)
ax1.contourf(data1_xx, data1_yy, data1_zz, cmap='Paired')
ax1.set_ylabel("x2")
ax1.set_xlabel("x1")
ax1.set_xticks(())
ax1.set_yticks(())
ax1.set_title("Data 1")
# data2
data1_graph = plt.pyplot.figure(2)
ax2.contourf(data2_xx, data2_yy, data2_zz, cmap='Paired')
ax2.set_ylabel("x2")
ax2.set_xlabel("x1")
ax2.set_xticks(())
ax2.set_yticks(())
ax2.set_title("Data 2")
# data3
data1_graph = plt.pyplot.figure(3)
ax3.contourf(data3_xx, data3_yy, data3_zz, cmap='Paired')
ax3.set_ylabel("x2")
ax3.set_xlabel("x1")
ax3.set_xticks(())
ax3.set_yticks(())
ax3.set_title("Data 3")
```
